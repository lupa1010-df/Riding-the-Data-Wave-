```python
import pandas as pd
import numpy as np
```

```python
# Newark Airport 2016 Data 

# Load the Data 

file_path = "/home/pauli_bi/Projects/02_City_Bike_Ridership/bike-rental-starter-kit/bike-rental-starter-kit/data"
df_new_air = pd.read_csv(file_path + "/newark_airport_2016.csv")


# Inspect the Data

print("Newark Airport 2016 Data Inspection:")

print(df_new_air.describe())
print(df_new_air.info())
print(df_new_air.isna().sum())
df_new_air.head()
```

```python
# Newark Airport 2016 Data 

# Drop empty columns:
# PGTM = Peak Gust Time (time of peak wind gust, TSUN = Sunshine Duration (total amount of sunshine during the day, usually in hours)
# Station and Name are always the same, so no necessary information

df_new_air = df_new_air.drop(["PGTM","TSUN","STATION","NAME"], axis = 1)
```

```python
# Newark Airport 2016 Data 

# Adjust data types
# Fill NaN values with a specific integer, e.g., 0

df_new_air['WDF5'] = df_new_air['WDF5'].fillna(0).astype(int)
df_new_air['WSF5'] = df_new_air['WSF5'].fillna(0)

```

```python
# Citibike Tripdata 2016 (January - December)

# Load the Data 

df_jan = pd.read_csv(file_path + "/JC-201601-citibike-tripdata.csv")
df_feb = pd.read_csv(file_path + "/JC-201602-citibike-tripdata.csv")
df_mar = pd.read_csv(file_path + "/JC-201603-citibike-tripdata.csv")
df_apr = pd.read_csv(file_path + "/JC-201604-citibike-tripdata.csv")
df_may = pd.read_csv(file_path + "/JC-201605-citibike-tripdata.csv")
df_jun = pd.read_csv(file_path + "/JC-201606-citibike-tripdata.csv")
df_jul = pd.read_csv(file_path + "/JC-201607-citibike-tripdata.csv")
df_aug = pd.read_csv(file_path + "/JC-201608-citibike-tripdata.csv")
df_sep = pd.read_csv(file_path + "/JC-201609-citibike-tripdata.csv")
df_oct = pd.read_csv(file_path + "/JC-201610-citibike-tripdata.csv")
df_nov = pd.read_csv(file_path + "/JC-201611-citibike-tripdata.csv")
df_dec = pd.read_csv(file_path + "/JC-201612-citibike-tripdata.csv")

#Concatenate the Dataframes

df_cit_trip = pd.concat([df_jan, df_feb, df_mar, df_apr, df_may, df_jun, df_jul, df_aug, df_sep, df_oct, df_nov, df_dec])

# Inspect the Data

print("Citibike Tripdata Inspection:")
print(df_cit_trip.describe())
print(df_cit_trip.info())
print(df_cit_trip.isna().sum())
```

```python
# Citibike Tripdata 2016 (January - December)

# Inspect the missing Values of 'User Type' and 'Birth Year'

missing_birth_year = df_cit_trip[df_cit_trip['Birth Year'].isna()]

print('Missing Birth Year')
print(missing_birth_year.describe())
print(missing_birth_year.info())

# The missing values of the column 'Birth Year' are replaced by 0. These rows also have a missing value in 'Gender' saved as 0.
# The variable type of the column is changed from float to integer

df_cit_trip['Birth Year'] = df_cit_trip['Birth Year'].fillna(0).astype(int)

# Inspect the missing Values of 'User Type'

missing_user_type = df_cit_trip[df_cit_trip['User Type'].isna()]

print('Missing User Type')
print(missing_user_type.describe())
print(missing_user_type.info())

# The column 'User Type' remains untouched and the missing values are not removed/ altered
```

```python
# Citibike Tripdata 2016 (January - December)


# Create a new Column 'Date', which can be later useful to join with the Newark Airport 2016 Data 

start_time = pd.to_datetime(df_cit_trip['Start Time'])
df_cit_trip['Date'] = start_time.dt.strftime('%Y-%m-%d')
```

```python
# Create the relational Databases in Python (Take a look at the schema.)

# Create the Stations DataFrame
# Select the columns and rename them
end_stations_df = df_cit_trip[['End Station ID', 'End Station Name', 'End Station Latitude', 'End Station Longitude']]
start_stations_df = df_cit_trip[['Start Station ID', 'Start Station Name', 'Start Station Latitude', 'Start Station Longitude']]

# Rename the columns
end_stations_df = end_stations_df.rename(columns={
    'End Station ID': 'station_id',
    'End Station Name': 'station_name',
    'End Station Latitude': 'station_latitude',
    'End Station Longitude': 'station_longitude'
})
start_stations_df = start_stations_df.rename(columns={
    'Start Station ID': 'station_id',
    'Start Station Name': 'station_name',
    'Start Station Latitude': 'station_latitude',
    'Start Station Longitude': 'station_longitude'
})

# Concatenate the DataFrames

stations_df = pd.concat([start_stations_df, end_stations_df], axis = 0)

# Drop the duplicate rows

stations_df = stations_df.drop_duplicates()

# Display the DataFrame

stations_df.head()
```

```python
# Create the relational Databases in Python (Take a look at the schema.)

# Create the Users DataFrame
users_df = df_cit_trip[['Birth Year', 'Gender', 'User Type']].copy()  # Make a copy

# Create a unique 'user_id' as the index
users_df['user_id'] = range(1, len(users_df) + 1)

# Set 'user_id' as the index
#users_df = users_df.set_index('user_id')

# Rename the columns
users_df = users_df.rename(columns={
    'Birth Year': 'birth_year',
    'Gender': 'gender',
    'User Type': 'user_type'
})

users_df.head()
```

```python
# Create the Dates DataFrame by copying the 'DATE' column
dates_df = df_new_air[['DATE']].copy()  # Ensure it's a DataFrame (not just a Series)

# Create a unique 'date_id' as the index
dates_df['date_id'] = range(1, len(dates_df) + 1)

# Set 'date_id' as the index
#dates_df = dates_df.set_index('date_id')

# Ensure that 'DATE' is in the right format
dates_df['DATE'] = pd.to_datetime(dates_df['DATE'])

# Rename the columns
dates_df = dates_df.rename(columns={
    'DATE': 'date'
})

# Display the DataFrame
dates_df.head()
```

```python
# Create the Trips DataFrame by selecting the relevant columns and making a copy
trips_df = df_cit_trip[['Trip Duration', 'Start Time', 'Stop Time', 'Start Station ID', 'End Station ID', 'Bike ID', 'Date']].copy()

# Add the column user_id from users_df
trips_df['user_id'] = users_df['user_id']

# Ensure that the columns are in the right format
trips_df['Date'] = pd.to_datetime(trips_df['Date'])
trips_df['Start Time'] = pd.to_datetime(trips_df['Start Time'])
trips_df['Stop Time'] = pd.to_datetime(trips_df['Stop Time'])

# Merge date_id from dates_df
trips_df = pd.merge(trips_df, dates_df, left_on='Date', right_on='date', how='left')

# Create a unique 'trip_id' as a column
trips_df['trip_id'] = range(1, len(trips_df) + 1)

# Drop the 'Date' and 'date' column
trips_df = trips_df.drop(['Date','date'], axis = 1)

# Rename the columns
trips_df = trips_df.rename(columns={
    'Trip Duration': 'trip_duration',
    'Start Time': 'start_time',
    'Stop Time': 'stop_time',
    'Start Station ID': 'start_station_id',
    'End Station ID': 'end_station_id',
    'Bike ID': 'bike_id'
})

# Display the DataFrame
trips_df.head()
```

```python
# Create the relational Databases in Python (Take a look at the schema.)

# Create the Weather DataFrame
#weather_df = df_new_air[['weather_id', 'date_id', 'awnd', 'prcp', 'snow', 'snwd', 'tavg', 'tmax', 'tmin','wdf2', 'wdf5', 'wsf2', 'wsf5']]
weather_df = df_new_air[['AWND', 'PRCP', 'SNOW', 'SNWD', 'TAVG', 'TMAX', 'TMIN','WDF2', 'WDF5', 'WSF2', 'WSF5']]

#Rename the colums
weather_df.columns = weather_df.columns.str.lower()

# Create a unique 'weather_id' as a column
weather_df['weather_id'] = range(1, len(weather_df) + 1)

# Create a unique 'date_id' as a column
weather_df['date_id'] = range(1, len(weather_df) + 1)

weather_df.head()
```

```python
# Exporting the DataFrames
weather_df.to_csv('weather.csv', index=False)
trips_df.to_csv('trips.csv', index=False)
dates_df.to_csv('dates.csv', index=False)
users_df.to_csv('users.csv', index=False)
stations_df.to_csv('stations.csv', index=False)
```

```python
# Connect with local Database
from sqlalchemy import create_engine
import logging

# Database connection parameters
username = 'postgres'
password = '********'
host = 'localhost'
port = '5432'
database = 'bike_rental'

# Create the connection string and engine
connection_string = f'postgresql://{username}:{password}@{host}:{port}/{database}'
engine = create_engine(connection_string)

# Inserting data from Pandas DataFrames into PostgreSQL tables
with engine.begin():  # This manages transactions automatically
    dates_df.to_sql('Dates', engine, if_exists='replace', index=False)
    stations_df.to_sql('Stations', engine, if_exists='replace', index=False)
    users_df.to_sql('Users', engine, if_exists='replace', index=False)
    weather_df.to_sql('Weather', engine, if_exists='replace', index=False)
    trips_df.to_sql('Trips', engine, if_exists='replace', index=False)

# Verify if data is inserted
print("Data has been successfully inserted into PostgreSQL!")
```
